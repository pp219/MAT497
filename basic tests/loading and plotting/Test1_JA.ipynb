```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from scipy.interpolate import interp2d

x=plot['x']
y=plot['y']
z=plot['z']

contour_breaks = 20
ticks = np.linspace(x.min(), x.max(), contour_breaks, endpoint=True)

fig = plt.figure()
axes = fig.add_subplot(111,aspect='equal')
f = interp2d(x,y,z,kind='cubic')
z1 = f(x,y)
#axes.contour(x,y,z1,colors='k')
fill = axes.contourf(x,y,z1,ticks,cmap=cm.viridis)
fig.colorbar(fill, ticks=ticks)

plt.show()
fig.savefig("./test2.png")
```

```python
grid_x, grid_y = np.meshgrid(np.linspace(x.min(), x.max(),
                                         np.int(np.floor(10000*(x.max()-x.min())))),
                             np.linspace(y.min(), y.max(),
                                         np.int(np.floor(10000*(y.max()-y.min())))))
```

```python
from scipy.interpolate import griddata
data = griddata((x,y),z,(grid_x,grid_y), method='cubic')
```

```python
plt.imshow(data, extent=[x.min(), x.max(),y.min(), y.max()])
plt.scatter(x,y,c='k',s=.2)
```
